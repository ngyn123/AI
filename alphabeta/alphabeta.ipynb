```python
import math
import random
from typing import List, Tuple, Optional

class TicTacToe:
    """
    Core game logic for Tic-Tac-Toe.
    Bot plays as 'X' (Bothicc), Human plays as 'O'.
    """
    BOT: str = "X"
    HUMAN: str = "O"
    EMPTY: str = "_"

    def __init__(self) -> None:
        self.board: List[List[str]] = [[self.EMPTY for _ in range(3)] for _ in range(3)]
        self.bot_turn: bool = random.choice([True, False])

    def print_board(self) -> None:
        """Hiển thị bàn cờ ra console."""
        print("\n  0 1 2")
        for idx, row in enumerate(self.board):
            print(f"{idx} " + "|".join(row))
        print()

    def get_valid_moves(self) -> List[Tuple[int, int]]:
        """Lấy danh sách các ô còn trống."""
        return [(i, j) for i in range(3) for j in range(3) if self.board[i][j] == self.EMPTY]

    def make_move(self, i: int, j: int, player: str) -> bool:
        """Thực hiện nước đi nếu hợp lệ."""
        if 0 <= i < 3 and 0 <= j < 3 and self.board[i][j] == self.EMPTY:
            self.board[i][j] = player
            return True
        return False

    def check_winner(self) -> Optional[str]:
        """Kiểm tra trạng thái thắng/thua. Trả về 'X', 'O' hoặc None."""
        b = self.board
        # Kiểm tra Hàng ngang và Cột dọc
        for i in range(3):
            if b[i][0] == b[i][1] == b[i][2] != self.EMPTY: return b[i][0]
            if b[0][i] == b[1][i] == b[2][i] != self.EMPTY: return b[0][i]

        # Kiểm tra Đường chéo
        if b[0][0] == b[1][1] == b[2][2] != self.EMPTY: return b[0][0]
        if b[0][2] == b[1][1] == b[2][0] != self.EMPTY: return b[0][2]

        return None

    def is_draw(self) -> bool:
        """Kiểm tra trạng thái hòa."""
        return all(cell != self.EMPTY for row in self.board for cell in row)

    def _minimax(self, depth: int, alpha: float, beta: float, is_maximizing: bool) -> float:
        """Thuật toán lõi Minimax kết hợp Alpha-Beta Pruning (Gộp Max và Min)."""
        winner = self.check_winner()
        if winner == self.BOT: return 10 - depth  # Thắng càng nhanh càng tốt
        if winner == self.HUMAN: return depth - 10
        if self.is_draw(): return 0

        moves = self.get_valid_moves()

        if is_maximizing:
            max_eval = -math.inf
            for (i, j) in moves:
                self.board[i][j] = self.BOT
                eval = self._minimax(depth + 1, alpha, beta, False)
                self.board[i][j] = self.EMPTY
                max_eval = max(max_eval, eval)
                alpha = max(alpha, eval)
                if beta <= alpha:
                    break
            return max_eval
        else:
            min_eval = math.inf
            for (i, j) in moves:
                self.board[i][j] = self.HUMAN
                eval = self._minimax(depth + 1, alpha, beta, True)
                self.board[i][j] = self.EMPTY
                min_eval = min(min_eval, eval)
                beta = min(beta, eval)
                if beta <= alpha:
                    break
            return min_eval

    def get_best_bot_move(self) -> Tuple[int, int]:
        """Tính toán nước đi tối ưu nhất cho Bot."""
        best_val = -math.inf
        best_move = (-1, -1)

        for (i, j) in self.get_valid_moves():
            self.board[i][j] = self.BOT
            move_val = self._minimax(0, -math.inf, math.inf, False)
            self.board[i][j] = self.EMPTY

            if move_val > best_val:
                best_move = (i, j)
                best_val = move_val

        return best_move

    def play(self) -> None:
        """Vòng lặp chính quản lý luồng game và tương tác UI."""
        print("="*40)
        print("Welcome to the deepest hell of TicTacToe")
        print("The thicc demon, Bothicc (X), will be your opponent!")
        print("FIGHT AND BE VICTORIOUS, SOLDIER!")
        print("="*40)

        while True:
            self.print_board()

            if self.check_winner() or self.is_draw():
                break

            if self.bot_turn:
                print("[BEHOLD, PEASANT! Bothicc is thinking...]")
                i, j = self.get_best_bot_move()
                self.make_move(i, j, self.BOT)
                self.bot_turn = False
            else:
                print("[MAKE YOUR MOVE!]")
                while True:
                    try:
                        i = int(input("Enter row index (0-2): "))
                        j = int(input("Enter column index (0-2): "))
                        if self.make_move(i, j, self.HUMAN):
                            break
                        print("[Invalid Move! Cell is taken or out of bounds. Try again.]")
                    except ValueError:
                        print("[Can't you differentiate numbers and literals!? Try again.]")
                self.bot_turn = True

        # End game sequence
        winner = self.check_winner()
        if winner == self.BOT:
            print("\n[Bothicc is superior! YOU LOSE!]")
        elif winner == self.HUMAN:
            print("\n[You're just lucky! YOU WIN!]")
        else:
            print("\n[It's a Tie! Bothicc is still superior!]")

        self.retry()

    def retry(self) -> None:
        """Reset game theo lựa chọn của người chơi."""
        print("\n[Wanna try again, LOSER?]")
        choice = input("Enter 'NO' to quit, or anything else to continue: ").strip().lower()
        if choice != "no":
            self.__init__()  # Reset trạng thái
            self.play()
        else:
            print("\n[Now now, don't cry. Goodbye!]")

if __name__ == "__main__":
    game = TicTacToe()
    game.play()
```
